```python
import numpy as np
import pandas as pd

import matplotlib as mlp
import matplotlib.pyplot as plt
```

**Question 1**

```python
##Plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)

plt.plot(a,b,'go--')
plt.xlabel('a')
plt.ylabel('b')
plt.title('Graph representing the relation between a and b',color='r')
```

**Question 2**

```python
# Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2


plt.plot(days,sales_1,'bo--',label ='Company 1')
plt.plot(days,sales_2,'ro--',label='Company 2')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend(loc ='lower right')
plt.title('Graph showing the sales trend in company 1 and 2',color='g')
```

**Question 3**

```python
# Create a 3 by 3 subplots:

## multiple plots

x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]


plt.subplot(3,3,1)
plt.plot(x,y1,'bo--')
plt.xlabel('x')
plt.ylabel('y1')

plt.subplot(3,3,2)
plt.plot(x,y2,'bo--')
plt.xlabel('x')
plt.ylabel('y2')

plt.subplot(3,3,3)
plt.plot(x,y3,'bo--')
plt.xlabel('x')
plt.ylabel('y3')

plt.subplot(3,3,4)
plt.plot(x,y4,'bo--')
plt.xlabel('x')
plt.ylabel('y4')

plt.subplot(3,3,5)
plt.plot(x,y5,'bo--')
plt.xlabel('x')
plt.ylabel('y5')

plt.suptitle('Subplots:', fontsize =17,color='r')
```
